That means, the position embedding layer of the transformers has 512 weights, but the sentence transformer will only use and was trained with the first 256 of them. Therefore, you should be careful with increasing the value above 256. It will work from a technical perspective, but the position embedding weights (>256) are not properly trained and can therefore mess up your results. Please also check this StackOverflow post.

```python
import pandas as pd
from pathlib import Path
from typing import List

def combine_and_save_parquet(parquet_paths: List[Path], output_path: Path) -> pd.DataFrame:
    # List to hold the DataFrames from each parquet file
    df_list = []
    
    # Iterate through the list of parquet paths
    for path in parquet_paths:
        # Load the parquet file into a DataFrame
        df = pd.read_parquet(path)
        # Append the DataFrame to the list
        df_list.append(df)
    
    # Concatenate all DataFrames into a single DataFrame
    combined_df = pd.concat(df_list, ignore_index=True)
    
    # Save the combined DataFrame as a parquet file
    combined_df.to_parquet(output_path, index=False)
    
    return combined_df

# Example usage
parquet_paths = [Path("datasets/literature_train.parquet"), Path("datasets/wiki_train.parquet")] 
                #  Path("datasets/news_train.parquet"), Path("datasets/science_train.parquet")]
output_path = Path("datasets/train.parquet")
combined_df = combine_and_save_parquet(parquet_paths, output_path)
```
