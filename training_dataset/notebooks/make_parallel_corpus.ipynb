```python
# Specify the path to your txt file
file_path = Path("../datasets/MaCoCu-sr-en.latin.sent.txt")

# Load the file into a pandas DataFrame
with open(file_path, 'r', encoding='utf-8') as file:
    header = file.readline()
    expected_num_cols = len(header.split('\t'))  # Count the columns in the header

# Function to filter out rows with the wrong number of fields
def load_correct_rows(file_path, expected_num_cols):
    correct_rows = []
    with open(file_path, 'r', encoding='utf-8') as file:
        header = file.readline()  # Skip the header
        for line in file:
            if len(line.split('\t')) == expected_num_cols:
                correct_rows.append(line.strip().split('\t'))
    
    # Create DataFrame from correct rows
    df = pd.DataFrame(correct_rows, columns=header.strip().split('\t'))
    return df

# Load only the correct rows
df_correct = load_correct_rows(file_path, expected_num_cols)

# Display the first few rows to ensure it loaded correctly
print(df_correct.head())
```

```python
data_dict = df[["sentence1", "sentence2"]].rename(columns={"sentence1": "english", "sentence2": "non_english"})
dataset = Dataset.from_pandas(data_dict)
```

```python
from sentence_transformers import SentenceTransformer


teacher_model = SentenceTransformer('intfloat/e5-base-v2')
inference_batch_size = 64
def encode_sentences(examples):
    # Encode all sentences in the "english" column in batches
    encoded = teacher_model.encode(
        examples["english"],
        batch_size=inference_batch_size,
        # show_progress_bar=True,  # Show a progress bar for feedback
    )
    return {"label": encoded}  # Add encoded representations to the dataset
small_dataset = dataset.select(range(70))

# Use the Hugging Face map function to process the dataset efficiently
train_dataset = small_dataset.map(encode_sentences, batched=True, batch_size=inference_batch_size)
```
